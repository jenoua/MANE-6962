```python
#Problem 2 ========================================================================

import tensorflow as tf
from numpy import loadtxt
from keras import optimizers
from tensorflow.python.keras.optimizers import *
import numpy as np
import pandas as pd

data = pd.read_csv(r'NACA0012 Data.csv', dtype=float)
data = data.to_numpy()

#Visualize the Data ______________________________________
fig, axes = plt.subplots(2,3, figsize=(20,10))
labels = ["Frequency","AoA","Chord Length","Free-Stream Velocity","Displacement"]
for i, ax in enumerate(axes.flatten()):
 if i > 4:
  break

 ax.scatter(data[:,i], data[:,5])
 ax.set_xlabel(labels[i])
 ax.set_ylabel('Scaled Sound Pressure Level')
    
for i in range(3):
 data[:,i] = (data[:,i] - data[:,i].mean(axis=0)) / data[:,i].std(axis=0)

#Algorithm ___________________________________________
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test=train_test_split(
    data[:,0:4],data[:,5],
    test_size=0.30,
    random_state=123,
    shuffle=True)

model = Sequential()
model.add(Dense(256, activation='relu', input_dim=4))
model.add(Dense(128, activation='relu'))
model.add(Dense(64, activation='relu'))
model.add(Dense(1))

mse = tf.keras.losses.MeanSquaredError(reduction=tf.keras.losses.Reduction.SUM)
model.compile(optimizer='adam',loss=mse)

model.summary()
history = model.fit(X_train, y_train, epochs=100, batch_size=5, validation_split=0.3)
                    

#Plot Validation Loss
fig, ax = plt.subplots(figsize=(4,4))
ax.plot(history.history['loss'])
ax.plot(history.history['val_loss'])
ax.set_title('Model Loss')
ax.set_ylabel('Loss')
ax.set_xlabel('Epoch')
plt.legend(['Train', 'Validation'])
ax.set_yscale('log')
plt.show()

#Coefficient of Determination _________________________________________________
from sklearn.metrics import r2_score
y_train_pred = model.predict(X_train).flatten()
r2 = r2_score(y_train, y_train_pred)
print("R^2 Score: ", r2)

print("The model yielded an R2 value of 0.68 and followed a suitable trend for validation compared to training.")
print("I believe the model can be improved by changing the combination of epoch, batch size, or model architecture (# of neurons, # of layers).")
print("Using a Machine Learning model allowed us to quickly find patterns for identifying scaled sound pressure levels")
print("This saves time and can lead to fewer mistakes")
```
